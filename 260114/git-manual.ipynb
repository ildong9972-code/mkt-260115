# git 사용방법
 1) 로컬컴퓨터(내 컴퓨터)에 git 응용프로그램 설치
 2) 터미널 : git -v 명령어
    - 검색어 > cmd (명령프롬프트) 입력 후 클릭 > 검정색 창 (윈도우)
 3) 만약, 없다면 > 구글 크롬 > git 다운로드 후 설치 (https://git-scm.com/)
    - 대리로 응용프로그램 다운로드 후 설치해주는 중간 프로그램: chocolaty (윈도우) (https://chocolatey.org/)
 4) 깃의 소유주 확인 & 변경
    - git config --list
 5) github 회원가입 : 계정정보 값으로 입력 (github에서는 웬만하면 수정하지 마시길...)
    - git : 나의 로컬컴퓨터에서 버전 관리 툴
    - github : 나의 로컬컴퓨터에서 관리하고 있는 버전을 타인과 공유해서 함께 사용하고자 할 때, 혹은 내가 사용하고 있는 디바이스가 2대 이상일 경우, A에서도 데이터 관리, B에서도 데이터 관리하고자 할 때 (https://github.com/)
    - repository : github > project에 따른 개별 폴더 (개수 제한 없음)
    - user.email : ildong9972@gmail.com
    - user.name : ildong9972-code
    - 터미널 재접속 > 관리자 모드로 접속 = 전역(global) 세팅
    - git config --global user.name "ildong9972-code"
    - git config --global user.email "ildong9972@gmail.com"
 6) git commit & push
    - cd "현재 나의 루트폴더"까지 이동
    - git 초기화하기 (git init)
    - staging area : 데이터 전송 : git add .
    - 데이터 tracking 이 정상적으로 됐는지 확인하는 법: git status
    - 데이터 labeling : git commit -m "260115"
    - 원격 (웹) github : git remote add origin 레포지토리 주소 (나의 오리지널 레포지토리 주소를 git에 연결)
    - 로컬(git) & 원격(github) 잘 연결되었을까? -> git remote -v fetch/push (받기/보내기)
    - 로컬 -> 원격으로 데이터 푸쉬 : git push origin master
 7) git clone & pull
    - 집 도착 > 컴퓨터 켬 > 집 컴 git 설치 > 새 폴더 생성 (home-pc)해서 주소창의 주소 복사 > 터미널로 ㄱㄱ
    - cd /(주소복붙) > 엔터
    - ~~ git clone
    - 새로운 폴더(Root 폴더)
    - 해당 루트 폴더 : git clone https://github.com/ildong9972-code/mkt-260115 > 엔터
    - home-pc 폴더 확인 > mkt-260115 폴더 생성된 거 확인
    - 집에서 복습, 학원파일을 오픈해서 추가로 내용 업데이트 > 다시 내 레포지토리 푸쉬
    - git add . > 해당 명령어부터 시작!!! git clone에서 내려 받는다 -> 쓴다 -> git add . 로 사용해서 올린다. (git init 쓰면 안됨)
 8) git fork (누군가의 레포지토리를 떠서 내 쪽으로 가져올 수 있음 - 업뎃 될 때마다 클론, 복제해서 가져오면 됨)
