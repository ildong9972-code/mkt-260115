# 데이터 자료 구조 1 - 리스트 (list) 
- 문자열, 정수, 실수, 논리형 -> (변수 : 값) // 300개가 필요할 경우, 변수 300개 필요
- 변수명 하나: 여러가지의 값을 그룹화
  - 공통분모를 가지고 있는 값들의 모임 -> 특성 분류
  - 상대적으로 컴퓨터 메모리 공간 효율적 사용
  - 대부분의 크롤링한 결과값 -> 리스트 형태의 자료 구조 값이 결정됨
- 변수명 = [값, 값, 값 ]
- 리스트 & 문자열 : 유사성을 띄고 있다 -> 비슷한 문법 구조를 가지고 있다
- 리스트, 문자열 -> "공통속성"을 가지고 있기 때문에 위와 같은 이야기가 나오는 것임!
- __iter__ -> iterable : 반복할 수 있는 / index
- 값을 추가하는 방법
  - append() : 리스트에서 값을 추가 (뒤에서)
  - 리스트 + 리스트 연산
  - insert(추가하고자 하는 인덱스, 넣고자 하는 값)
  - extend(여러 개의 값을 한 번에 넣고자 할 때)
- 값을 제거하는 밥법
  - pop() : 리스트에 값을 제거 (뒤에서) (append의 반대 개념) -> 인자값으로 인덱스값 부여 -> 해당 인덱스 값 제거됨
  - remove() : 리스트에서 특정 값을 직접 선택해서 제거
  - del 리스트명[인덱스] : 해당 인덱스의 값을 제거
- 슬라이싱 기능 = 문자열과 동일 작동
- 리스트 자료 구조 안에 입력될 값의 타입이 상이해도 가능

```python
location = ["서울시", "경기도", "인천시"]
print(location)
print(type(location))
print(location[0])
print(location[1])
print(location[2])
print(location[-1])
```
