# 파이썬 기초문법
- 출력 : print() -> 내장함수 = built in function / 내장: 굳이 우리가 만들지 않아도 애초에 파이썬 문법 탄생될 때부터 존재.. 
  사칙연산 등도 가능
- 입력 : input("당신의 이름은?") -> 사용자로부터 어떤 값을 입력받을 수 있음 / placeholder = 입력해야할 값의 정보
         해당 함수를 통해 수집한 값은 무조건 "문자열(str)"이 된다는 사실 기억!!!
- 연산 : 사칙연산 : + @ -> +, -, X, /, % (나머지), ** (제곱근) -> '+' : 문자 간의 연결연산자로서도 사용 가능!
         비교연산 : < > <= >= == != (!= 부정연산자... 같지 않다는 뜻) ("좌항 = 우항, 우항의 값을 좌항으로 삽입, 적용)  
- 변수 - variable -> 변할 수 있는 값을 넣는 공간 = 그릇
       - 변수명 작명 => 직관적이고 명시적이어야 함 -> 누가 봐도 한눈에 알아볼 수 있는 변수명. 숫자가 먼저 올 수 없음 (중간에 오는 건 가능)
       - 팁 => 스네이크 표기법 (뱀 -> case_study / 카멜 표기법 (낙타 -> caseStudy) 팀플 시 왠만하면 같은 표기법 사용
- 자료형 : 변수 안에 담을 수 있는 값을 형태. 총 네 가지 -> 문자열 (String) / 정수 (Int) / 실수 (float) / 논리형 (Boolean)
  소수점을 가지고 있는 숫자 ㅡ 실수 / 다른 프로그램들은 문자열과 문자를 구분함 ("", '' 구분 ㅡ 파이썬은 구분 안 함)
           형변화: 자동형변환 (input 함수를 거치면서 자동으로 형태 변환) / 명시적형변환 (수동으로 형태 변환)
           자료형확인 : type()
           int() : 정수형 변환
           float() : 실수형 변환
           str() : 문자형 변환
           bool() : 논리형 변환
           list() : 리스트 자료구조 변환
           dict() : 딕셔너리 자료구조 변환
           tuple() : 튜플 자료구조 변환
           set() : 세트(집합) 자료구조 변환
  - 문자열 : 독립적인 함수 혹은 속성들이 많이 존재
             - count() : 특정 문자열에서 어떤 단어가 몇 번 사용되었는지를 집계하고자 할 때 (대,소문자 구분) ㅡ 경쟁사 분석 시 사용 가능 ex) 경쟁사 웹 크롤링 후 리뷰 긍정/부정 단어 분석
             - len() : length = 길이. 띄어쓰기도 문자로 들어감 (댓글의 길이값 찾을 수 있음)
             - find() : 해당 문자열에서 인자값으로 전달된 문자가 몇 번째 인덱스에 위치하는지 찾고자 할 때 사용 (인덱스 넘버는 0부터 시작)
             - replace() : 특정 문자열을 대체하고자 할 때
             - index : 문자열 안에 각각의 문자는 고유한 숫자를 가지고 있다 -> 문자열 왼쪽부터 숫자 시작은 무조건 0부터 시작
             - slicing : 문자열 안에서 특정 구간을 잘라내서 가져오고 싶을 때 (자주 씀)
             - strip() : 특정 문자열 앞,뒤쪽에 여백 혹은 인지할 수 있는 escape sequence (\n, \t) 문자열을 제거하고자 할 때 사용
             - format() : 특정 문자열을 출력하고자 할 때, 사용할 수 있는 메서드 함수(기능)
    - 형식지정자 :
             - 특정 타입의 값을 출력하고자 할 때, 어떤 형태로 값을 찾아와서 출력할 것인지 결정
             - %s - string
             - %c - character
             - %d - int (정수)
             - %f - float (실수)

```python
# 반복문 사용

text = "banana"
target = "a"

idx = text.find(target) # 존재하지 않는 문자를 찾아오라고 하면 -1이 나옴

# 예외 조항 처리 : 어떤 명령문을 실행시키고자 할 때, 만약 ~가 아니라면, ~해라!

while idx != -1 :
    print(idx)
    idx = text.find(target, idx + 1)
```

```python
text = "python"

print(text)
print(text[2])
print(text[2:4]) # 찾아오고자 하는 문자열[start:finish+1]
print(text[2:6])
print(text[2:]) # 슬라이싱에서 마지막 번째 인덱스 생략 = 끝까지 의미
print(text[2:len(text)]) # 해당 구문을 많이 사용함
print(text[2:-1])
```

```python
name = "I have a {}, I have an {}."
print(name)

print(name.format("pen", "apple"))
print("I have a {}, I have an {}.".format("pen", "apple"))
print("I have an {1}, I have a {0}.".format("pen", "apple")) # 매서드(method) 함수 - 앞에 반드시 대상(객체)이 있어야 함
                                                             # 함수 앞에 .이 있으면 매서드 함수임
                                                             # 문자열 형식의 객체.format
```

```python
interest = 0.087
print(format(interest, ".2f")) # 독립함수 (문자, 숫자x) -> 소수점 위치 일치
```

```python
# 파이썬 내장함수 가운데, 소수점 자리를 올리거나 내리는 독립 기능은 없음
# 외부에서 라이브러리 모듈을 찾아와서 사용해야함
# 수학 객체 라이브러리 모듈

import math

interest = 0.891
result01 = math.floor(interest) # 바닥 -> 버리다
print(result01)

result02 = math.ceil(interest) # 천장 -> 올리다
print(result02)
```

```python
print("I have a %f, I have an %c." % (3.14, 'A')) # 엄격한 프로그래밍 언어 -> " ", ' '
print("I have a %f, I have an %c." % (3.14, 65)) # 정수 형식 (https://en.cppreference.com/w/cpp/language/ascii.html 에서 숫자 65는 A임을 확인 가능)
print("I have a %f, I have an %s." % (3.14, "AB")) # AB는 문자열이라 "" 처리
```
